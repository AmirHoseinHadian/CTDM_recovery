```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

import matplotlib
matplotlib.rcParams['font.serif'] = "Palatino"
matplotlib.rcParams['font.family'] = "serif"

plt.rc('axes', titlesize=16)     # fontsize of the axes title
plt.rc('axes', labelsize=16)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=12)    # fontsize of the tick labels
plt.rc('ytick', labelsize=12)    # fontsize of the tick labels
plt.rc('legend', fontsize=15)    # legend fontsize
```

```python
hyp_df = pd.read_csv('_01_IE/_02_ndt_constrain/_data/Hyperbolic_rec_500.csv')
exp_df = pd.read_csv('_01_IE/_02_ndt_constrain/_data/Exponential_rec_500.csv')

bhyp_df = pd.read_csv('_01_IE/_01_no_constrain/_data/Hyperbolic_rec_500.csv')
bexp_df = pd.read_csv('_01_IE/_01_no_constrain/_data/Exponential_rec_500.csv')
```

```python
plt.figure(figsize=(15, 6), layout='constrained')

plt.subplot(2, 5, 1)
plt.plot(np.arange(1.5, 4, 0.01), np.arange(1.5, 4, 0.01), c='red')
sns.scatterplot(x='a_true', y='a_estimate', data=hyp_df, alpha=0.2, color='black');
plt.xlabel(r'$\theta$ true')
plt.ylabel('Hyperbolic\n\n'+ r'$\theta$ estimate')

plt.subplot(2, 5, 2)
plt.plot(np.arange(0, 2, 0.01), np.arange(0, 2, 0.01), c='red')
sns.scatterplot(x='lambda_true', y='lambda_estimate', data=hyp_df, alpha=0.2, color='black');
plt.xlabel(r'$\lambda$ true')
plt.ylabel(r'$\lambda$ estimate')

plt.subplot(2, 5, 3)
plt.plot(np.arange(0, 3, 0.01), np.arange(0, 3, 0.01), c='red')
sns.scatterplot(x='delta_true', y='delta_estimate', data=hyp_df, alpha=0.2, color='black');
plt.xlabel(r'$\delta$ true')
plt.ylabel(r'$\delta$ estimate')

plt.subplot(2, 5, 4)
plt.plot(np.arange(0, 1, 0.01), np.arange(0, 1, 0.01), c='red')
sns.scatterplot(x='ndt_true', y='ndt_estimate', data=hyp_df, alpha=0.2, color='black');
plt.xlabel(r'$\tau$ true')
plt.ylabel(r'$\tau$ estimate')

plt.subplot(2, 5, 5)
plt.plot(np.arange(0, 1, 0.01), np.arange(0, 1, 0.01), c='red')
sns.scatterplot(x='sigma_true', y='sigma_estimate', data=hyp_df, alpha=0.2, color='black');
plt.xlabel(r'$\sigma$ true')
plt.ylabel(r'$\sigma$ estimate');

#############################

plt.subplot(2, 5, 6)
plt.plot(np.arange(1.5, 4, 0.01), np.arange(1.5, 4, 0.01), c='red')
sns.scatterplot(x='a_true', y='a_estimate', data=exp_df, alpha=0.2, color='black');
plt.xlabel(r'$\theta$ true')
plt.ylabel('Exponential\n\n' + r'$\theta$ estimate')

plt.subplot(2, 5, 7)
plt.plot(np.arange(0, 2, 0.01), np.arange(0, 2, 0.01), c='red')
sns.scatterplot(x='lambda_true', y='lambda_estimate', data=exp_df, alpha=0.2, color='black');
plt.xlabel(r'$\lambda$ true')
plt.ylabel(r'$\lambda$ estimate')

plt.subplot(2, 5, 8)
plt.plot(np.arange(0, 3, 0.01), np.arange(0, 3, 0.01), c='red')
sns.scatterplot(x='delta_true', y='delta_estimate', data=exp_df, alpha=0.2, color='black');
plt.xlabel(r'$\delta$ true')
plt.ylabel(r'$\delta$ estimate')

plt.subplot(2, 5, 9)
plt.plot(np.arange(0, 1, 0.01), np.arange(0, 1, 0.01), c='red')
sns.scatterplot(x='ndt_true', y='ndt_estimate', data=exp_df, alpha=0.2, color='black');
plt.xlabel(r'$\tau$ true')
plt.ylabel(r'$\tau$ estimate')

plt.subplot(2, 5, 10)
plt.plot(np.arange(0, 1, 0.01), np.arange(0, 1, 0.01), c='red')
sns.scatterplot(x='sigma_true', y='sigma_estimate', data=exp_df, alpha=0.2, color='black');
plt.xlabel(r'$\sigma$ true')
plt.ylabel(r'$\sigma$ estimate');

plt.savefig('plots/IE_recovery_plot.pdf', 
            transparent=True, 
            dpi=400, 
            bbox_inches='tight');
```
