```python
import numpy as np
import pandas as pd

from sklearn.metrics import r2_score

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
no_const_exp = pd.read_csv('_01_no_constrain/Exponential_rec.csv', index_col=0)
ndt_const_exp = pd.read_csv('_02_ndt_constrain/Exponential_rec.csv', index_col=0)
ndt_res_const_exp = pd.read_csv('_03_ndt_constrain_two_source/Exponential_rec.csv', index_col=0)

no_const_hyp = pd.read_csv('_01_no_constrain/Hyperbolic_rec.csv', index_col=0)
ndt_const_hyp = pd.read_csv('_02_ndt_constrain/Hyperbolic_rec.csv', index_col=0)
ndt_res_const_hyp = pd.read_csv('_03_ndt_constrain_two_source/Hyperbolic_rec.csv', index_col=0)
```

```python
df_exp = pd.DataFrame({'prms': ['theta', 'lambda'],
                       'DDM': [no_const_exp.corr()['a_estimate']['a_true'], 
                                    no_const_exp.corr()['lambda_estimate']['lambda_true']],
                       'ndt observation': [ndt_const_exp.corr()['a_estimate']['a_true'], 
                                     ndt_const_exp.corr()['lambda_estimate']['lambda_true']],
                       'Two source ndt observation': [ndt_res_const_exp.corr()['a_estimate']['a_true'], 
                                         ndt_res_const_exp.corr()['lambda_estimate']['lambda_true']]}).set_index('prms')

df_hyp = pd.DataFrame({'prms': ['theta', 'lambda'],
                       'DDM': [no_const_hyp.corr()['a_estimate']['a_true'], 
                                    no_const_hyp.corr()['lambda_estimate']['lambda_true']],
                       'ndt observation': [ndt_const_hyp.corr()['a_estimate']['a_true'], 
                                     ndt_const_hyp.corr()['lambda_estimate']['lambda_true']],
                       'Two source ndt observation': [ndt_res_const_hyp.corr()['a_estimate']['a_true'], 
                                         ndt_res_const_hyp.corr()['lambda_estimate']['lambda_true']]}).set_index('prms')
plt.figure(figsize=(8, 3))
plt.subplot(121)
plt.title('Exponential')
sns.heatmap(df_exp, vmin=0, vmax=1, annot=True, 
            cmap=sns.diverging_palette(15, 130, l=50, s=90, as_cmap=True));

plt.subplot(122)
plt.title('Hyperbolic')
sns.heatmap(df_hyp, vmin=0, vmax=1, annot=True, 
            cmap=sns.diverging_palette(15, 130, l=50, s=90, as_cmap=True));
```

```python
df_exp = pd.DataFrame({'prms': ['theta', 'lambda'],
                       'DDM': [r2_score(no_const_exp['a_estimate'], no_const_exp['a_true']),
                                    r2_score(no_const_exp['lambda_estimate'], no_const_exp['lambda_true'])],
                       'ndt observation': [r2_score(ndt_const_exp['a_estimate'], ndt_const_exp['a_true']),
                                     r2_score(ndt_const_exp['lambda_estimate'], ndt_const_exp['lambda_true'])],
                       'Two source ndt observation': [r2_score(ndt_res_const_exp['a_estimate'], ndt_res_const_exp['a_true']),
                                         r2_score(ndt_res_const_exp['lambda_estimate'], ndt_res_const_exp['lambda_true'])]}).set_index('prms')

df_hyp = pd.DataFrame({'prms': ['theta', 'lambda'],
                       'DDM': [r2_score(no_const_hyp['a_estimate'], no_const_hyp['a_true']),
                                    r2_score(no_const_hyp['lambda_estimate'], no_const_hyp['lambda_true'])],
                       'ndt observation': [r2_score(ndt_const_hyp['a_estimate'], ndt_const_hyp['a_true']),
                                     r2_score(ndt_const_hyp['lambda_estimate'], ndt_const_hyp['lambda_true'])],
                       'Two source ndt observation': [r2_score(ndt_res_const_hyp['a_estimate'], ndt_res_const_hyp['a_true']),
                                         r2_score(ndt_res_const_hyp['lambda_estimate'], ndt_res_const_hyp['lambda_true'])]}).set_index('prms')
plt.figure(figsize=(8, 3))
plt.subplot(121)
plt.title('Exponential')
sns.heatmap(df_exp, vmin=0, vmax=1, annot=True, 
            cmap=sns.diverging_palette(15, 130, l=50, s=90, as_cmap=True));

plt.subplot(122)
plt.title('Hyperbolic')
sns.heatmap(df_hyp, vmin=0, vmax=1, annot=True, 
            cmap=sns.diverging_palette(15, 130, l=50, s=90, as_cmap=True));
```
