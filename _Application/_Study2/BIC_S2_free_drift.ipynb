```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('CTDM_recovery/_Data/Study2.csv', 
                   index_col=0).reset_index(drop=True)

data = data.sort_values(by=['participant', 'trials', 'event'])

data = data[data['Duration']<data['RT']].reset_index(drop=True)

data['RT']/=1000
data['Duration']/=1000

data_speed = data[data.cue == 'SP'].reset_index(drop=True)
data_accuracy = data[data.cue == 'AC'].reset_index(drop=True)
```

```python
hyp_speed = pd.read_csv('CTDM_recovery/_Application/_Study2/_free_drift/_prms/hyp_speed.csv')
hyp_accuracy = pd.read_csv('CTDM_recovery/_Application/_Study2/_free_drift/_prms/hyp_accuracy.csv')
bhyp_speed = pd.read_csv('CTDM_recovery/_Application/_Study2/_free_drift/_prms/bhyp_speed.csv')
bhyp_accuracy = pd.read_csv('CTDM_recovery/_Application/_Study2/_free_drift/_prms/bhyp_accuracy.csv')
exp_speed = pd.read_csv('CTDM_recovery/_Application/_Study2/_free_drift/_prms/exp_speed.csv')
exp_accuracy = pd.read_csv('CTDM_recovery/_Application/_Study2/_free_drift/_prms/exp_accuracy.csv')
bexp_speed = pd.read_csv('CTDM_recovery/_Application/_Study2/_free_drift/_prms/bexp_speed.csv')
bexp_accuracy = pd.read_csv('CTDM_recovery/_Application/_Study2/_free_drift/_prms/bexp_accuracy.csv')
jddm_speed = pd.read_csv('CTDM_recovery/_Application/_Study2/_free_drift/_prms/jddm_speed.csv')
jddm_accuracy = pd.read_csv('CTDM_recovery/_Application/_Study2/_free_drift/_prms/jddm_accuracy.csv')
bddm_speed = pd.read_csv('CTDM_recovery/_Application/_Study2/_free_drift/_prms/bddm_speed.csv')
bddm_accuracy = pd.read_csv('CTDM_recovery/_Application/_Study2/_free_drift/_prms/bddm_accuracy.csv')
```

```python
def Behv_G2(cond_df, prms, N):
    cond_df['t0'] = -1
    cond_df['sig'] = -1

    for sbj in cond_df.participant.unique():
        cond_df.loc[cond_df.participant == sbj, 't0'] = prms.loc[prms.sbj == sbj, 't0'].values[0]
        cond_df.loc[cond_df.participant == sbj, 'sig'] = prms.loc[prms.sbj == sbj, 'sigma'].values[0]

    Z = cond_df.RT.values-cond_df.Duration.values
    tau = cond_df.t0.values
    sig = cond_df.sig.values
    obs_ll = 0.0005*((np.log(Z) - np.log(tau) - 0.5*sig**2)**2/(2*sig**2) + 0.5*np.log(2*np.pi * sig**2 * Z**2))

    return prms['G2'].sum() - 2*obs_ll.sum()
```

```python
print(bhyp_speed['G2'].sum() + bhyp_accuracy['G2'].sum() + 2*25*8)
print(bexp_speed['G2'].sum() + bexp_accuracy['G2'].sum() + 2*25*8)
print(bddm_speed['G2'].sum() + bddm_accuracy['G2'].sum() + 2*25*6)
```

```python
print(Behv_G2(data_speed, hyp_speed, 4) +
      Behv_G2(data_accuracy, hyp_accuracy, 4)+ 2*25*8)
print(Behv_G2(data_speed, exp_speed, 4) +
      Behv_G2(data_accuracy, exp_accuracy, 4)+ 2*25*8)
print(Behv_G2(data_speed, jddm_speed, 3)+
      Behv_G2(data_accuracy, jddm_accuracy, 3) + 2*25*6)
```
