```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('CTDM_recovery/_Data/Study2.csv', 
                   index_col=0).reset_index(drop=True)

data = data.sort_values(by=['participant', 'trials', 'event'])

data = data[data['Duration']<data['RT']].reset_index(drop=True)

data['RT']/=1000
data['Duration']/=1000
```

```python
hyp = pd.read_csv('CTDM_recovery/_Application/_Study2/_fixed_drift/_prms/hyp.csv')
bhyp = pd.read_csv('CTDM_recovery/_Application/_Study2/_fixed_drift/_prms/bhyp.csv')
exp = pd.read_csv('CTDM_recovery/_Application/_Study2/_fixed_drift/_prms/exp.csv')
bexp = pd.read_csv('CTDM_recovery/_Application/_Study2/_fixed_drift/_prms/bexp.csv')
jddm = pd.read_csv('CTDM_recovery/_Application/_Study2/_fixed_drift/_prms/jddm.csv')
bddm = pd.read_csv('CTDM_recovery/_Application/_Study2/_fixed_drift/_prms/bddm.csv')
```

```python
def Behv_G2(cond_df, prms, N):
    cond_df['t0'] = -1
    cond_df['sig'] = -1

    for sbj in cond_df.participant.unique():
        cond_df.loc[(cond_df.participant == sbj) & (cond_df.cue=='SP'), 't0'] = prms.loc[prms.sbj == sbj, 't0_sp'].values[0]
        cond_df.loc[(cond_df.participant == sbj) & (cond_df.cue=='AC'), 't0'] = prms.loc[prms.sbj == sbj, 't0_ac'].values[0]
        cond_df.loc[(cond_df.participant == sbj) & (cond_df.cue=='SP'), 'sig'] = prms.loc[prms.sbj == sbj, 'sigma_sp'].values[0]
        cond_df.loc[(cond_df.participant == sbj) & (cond_df.cue=='AC'), 'sig'] = prms.loc[prms.sbj == sbj, 'sigma_ac'].values[0]

    Z = cond_df.RT.values-cond_df.Duration.values
    tau = cond_df.t0.values
    sig = cond_df.sig.values
    obs_ll = ((np.log(Z) - np.log(tau) - 0.5*sig**2)**2/(2*sig**2) + 0.5*np.log(2*np.pi * sig**2 * Z**2))
    obs_ll[cond_df.cue=='AC'] *= 0.001
    return prms['G2'].sum() - 2*obs_ll.sum()
```

```python
print(bhyp['G2'].sum() + 2*25*7)
print(bexp['G2'].sum() + 2*25*7)
print(bddm['G2'].sum() + 2*25*5)
```

```python
print(Behv_G2(data, hyp, 7) + 2*25*7)
print(Behv_G2(data, exp, 7) + 2*25*7)
print(Behv_G2(data, jddm, 5) + 2*25*5)
```
