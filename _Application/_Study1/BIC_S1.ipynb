```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('CTDM_recovery/_Data/Study1.csv', 
                   index_col=0).reset_index(drop=True)

data = data[data.event == 3] # This event corresponds to decision time and the rest correspond to non-decision time
data = data[data['Duration']<data['rt']].reset_index(drop=True)

data['rt']/=1000
data['Duration']/=1000

data_speed = data[data.condition == 'speed'].reset_index(drop=True)
data_accuracy = data[data.condition == 'accuracy'].reset_index(drop=True)
```

```python
hyp_speed = pd.read_csv('CTDM_recovery/_Application/_Study1/_prms/hyp_speed.csv')
hyp_accuracy = pd.read_csv('CTDM_recovery/_Application/_Study1/_prms/hyp_accuracy.csv')
bhyp_speed = pd.read_csv('CTDM_recovery/_Application/_Study1/_prms/bhyp_speed.csv')
bhyp_accuracy = pd.read_csv('CTDM_recovery/_Application/_Study1/_prms/bhyp_accuracy.csv')
exp_speed = pd.read_csv('CTDM_recovery/_Application/_Study1/_prms/exp_speed.csv')
exp_accuracy = pd.read_csv('CTDM_recovery/_Application/_Study1/_prms/exp_accuracy.csv')
bexp_speed = pd.read_csv('CTDM_recovery/_Application/_Study1/_prms/bexp_speed.csv')
bexp_accuracy = pd.read_csv('CTDM_recovery/_Application/_Study1/_prms/bexp_accuracy.csv')
jddm_speed = pd.read_csv('CTDM_recovery/_Application/_Study1/_prms/jddm_speed.csv')
jddm_accuracy = pd.read_csv('CTDM_recovery/_Application/_Study1/_prms/jddm_accuracy.csv')
bddm_speed = pd.read_csv('CTDM_recovery/_Application/_Study1/_prms/bddm_speed.csv')
bddm_accuracy = pd.read_csv('CTDM_recovery/_Application/_Study1/_prms/bddm_accuracy.csv')
```

```python
def Behv_G2(cond_df, prms, N):
    cond_df['t0'] = -1
    cond_df['sig'] = -1

    for sbj in cond_df.participant.unique():
        cond_df.loc[cond_df.participant == sbj, 't0'] = prms.loc[prms.sbj == sbj, 't0'].values[0]
        cond_df.loc[cond_df.participant == sbj, 'sig'] = prms.loc[prms.sbj == sbj, 'sigma'].values[0]

    Z = cond_df.rt.values-cond_df.Duration.values
    tau = cond_df.t0.values
    sig = cond_df.sig.values
    obs_ll = (np.log(Z) - np.log(tau) - 0.5*sig**2)**2/(2*sig**2) + 0.5*np.log(2*np.pi * sig**2 * Z**2)

    return prms['G2'].sum() - 2*obs_ll.sum()
```

```python
print(Behv_G2(data_speed, hyp_speed, 5) + 2*25*5)
print(Behv_G2(data_speed, exp_speed, 5) + 2*25*5)
print(Behv_G2(data_speed, jddm_speed, 4) + 2*25*4)

print('=====')

print(Behv_G2(data_accuracy, hyp_accuracy, 5) + 2*25*5)
print(Behv_G2(data_accuracy, exp_accuracy, 5) + 2*25*5)
print(Behv_G2(data_accuracy, jddm_accuracy, 4) + 2*25*4)
```
